```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from pathlib import Path

project_root = Path.cwd().parent
csv_file = project_root / "data" / "satgpa.csv"
csv_path = str(csv_file)
df = pd.read_csv(csv_file)
```

```python
df.head()
```

```python
df.describe()
```

```python
# count zeros per column and total zeros
zeros_per_col = (df == 0).sum()
total_zeros = int(zeros_per_col.sum())
print("Zeros per column:\n", zeros_per_col)
print("Total zeros:", total_zeros)
```

```python
# Distribución de género
plt.hist(df["sex"], bins=30)
plt.xlabel("Género")
plt.ylabel("Frecuencia")
plt.title("Distribución del Género")
plt.show()
```

```python
fig, axes = plt.subplots(3, 2, figsize=(15, 20))

# First row
sns.boxplot(data=df, x="sex", y="sat_m", ax=axes[0, 0])
axes[0, 0].set_xlabel("Sex")
axes[0, 0].set_ylabel("Math SAT Score")
axes[0, 0].set_title("Math SAT Score by Sex")

sns.boxplot(data=df, x="sex", y="sat_v", ax=axes[0, 1])
axes[0, 1].set_xlabel("Sex")
axes[0, 1].set_ylabel("Verbal SAT Score")
axes[0, 1].set_title("Verbal SAT Score by Sex")

sns.boxplot(data=df, x="sex", y="sat_sum", ax=axes[1, 0])
axes[1, 0].set_xlabel("Sex")
axes[1, 0].set_ylabel("Total SAT Score")
axes[1, 0].set_title("Total SAT Score by Sex")

# Second row
sns.boxplot(data=df, x="sex", y="hs_gpa", ax=axes[1, 1])
axes[1, 1].set_xlabel("Sex")
axes[1, 1].set_ylabel("High School GPA")
axes[1, 1].set_title("High School GPA by Sex")

sns.boxplot(data=df, x="sex", y="fy_gpa", ax=axes[2, 0])
axes[2, 0].set_xlabel("Sex")
axes[2, 0].set_ylabel("First-Year GPA")
axes[2, 0].set_title("First-Year GPA by Sex")

axes[2, 1].axis("off")

plt.tight_layout()
plt.show()
```
